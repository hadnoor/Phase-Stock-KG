```python
node_map = {}
node_label_map, label_map, label_id = {}, {}, 1
for nodes in nodes_json:
    node = nodes['n']
    node_map[node['identity']] = node
    print(node)
    if node['labels'][0] not in label_map:
        label_map[node['labels'][0]] = label_id
        label_id += 1
    node_label_map[node['identity']] = label_map[node['labels'][0]]
```

```python
nifty_id = {}
INDEX = "nifty500" 

directory = "../../data/" + INDEX + "/"

ticker_list = {}
company_to_id = {}
company_id    = 0
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        ticker, name = filename.split("-")
        name = name[:-5]
        ticker_list[ticker] = name
        df = pd.read_csv(f)
        df = df.dropna()
        if df.shape[0] <= 2800:    # 13 years
            print("Skipping file: Less Training-Testing samples [{0} samples]".format(df.shape[0]))
            continue

        if ticker not in company_to_id:
            company_to_id[ticker] = company_id
            company_id += 1
nifty_id = company_to_id
print(len(nifty_id), nifty_id)
```
