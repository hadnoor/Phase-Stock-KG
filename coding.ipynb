```python

class DSU:
    def __init__(self, n):
        self.parent = list(range(n))
        self.size = [1] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.size[root_x] < self.size[root_y]:
                root_x, root_y = root_y, root_x
            self.parent[root_y] = root_x
            self.size[root_x] += self.size[root_y]

def find_total_groups(n, queryType, student1, student2):
    dsu = DSU(n)
    result = []
    for i in range(len(queryType)):
        if queryType[i] == 'Friend':
            dsu.union(student1[i] - 1, student2[i] - 1)
        elif queryType[i] == 'Total':
            group1 = dsu.find(student1[i] - 1)
            group2 = dsu.find(student2[i] - 1)
            result.append(dsu.size[group1] + dsu.size[group2])
    return result
# Example usage:
n = 4
queryType = ['Friend', 'Friend', 'Total']
student1 = [1, 2, 1]
student2 = [2, 3, 4]
print(find_total_groups(n, queryType, student1, student2)) 
```

```python
def calculate_averages(numbers):
   # Check if the length of numbers is exactly 18
   if len(numbers) != 18:
       raise ValueError("The input list must contain exactly 18 floating point numbers.")
   # Split the list into 6 sublists each containing 3 numbers
   lists_of_3 = [numbers[i:i + 3] for i in range(0, len(numbers), 3)]
   # Calculate the average of each sublist
   averages = [sum(sublist) / len(sublist) for sublist in lists_of_3]
   return averages
# Example usage
numbers = [0.551,0.42,0.071,0.623,0.482,0.049,0.611,0.517,0.031,0.545,0.48,0.065,0.49,0.36,0.76,0.586,0.68,0.56]
averages = calculate_averages(numbers)
print(averages)
def calculate_differences(numbers):
    # Check if the length of numbers is exactly 6
    if len(numbers) != 6:
        raise ValueError("The input list must contain exactly 6 numbers.")
    
    # The 6th number in the list
    sixth_number = numbers[5]
    
    # Calculate the difference between each number and the 6th number
    differences = [sixth_number - num  for num in numbers]
    
    return differences

# Example usage
# numbers = [10.5, 20.2, 30.3, 40.4, 50.5, 60.6]

differences = calculate_differences(averages)
print(differences)
def find_maximum(numbers):
    # Check if the length of numbers is exactly 6
    if len(numbers) != 6:
        raise ValueError("The input list must contain exactly 6 floating point numbers.")
    
    # Find the maximum element
    maximum = max(numbers)
    
    return maximum

# Example usage
# numbers = [10.5, 20.2, 30.3, 40.4, 50.5, 60.6]

maximum = find_maximum(differences)
print(f"The maximum element of the 6 numbers is: {maximum}")
```
